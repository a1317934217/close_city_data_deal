```python
import pandas as pd
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
```

# 读取数据

```python
data_01 = pd.read_csv('dataset_20210101.csv')
data_02 = pd.read_csv('dataset_20210102.csv')
data_03 = pd.read_csv('dataset_20210103.csv')
data_04 = pd.read_csv('dataset_20210104.csv')
data_08 = pd.read_csv('dataset_20210108.csv')
data_09 = pd.read_csv('dataset_20210109.csv')
data_10 = pd.read_csv('dataset_20210110.csv')
data_11 = pd.read_csv('dataset_20210111.csv')
data_12 = pd.read_csv('dataset_20210112.csv')
```

```python
data = pd.concat([data_01, data_02, data_03, data_04, data_08, data_09, data_10, data_11, data_12], ignore_index=True)
data
```

```python
data['hub_promoted_index'] = pd.to_numeric(data['hub_depressed_index'], errors='coerce')
```

# 下采样

```python
from sklearn.utils import resample
# 假设你有两个类别，分别为正类别和负类别
positive_samples = data[data['flag'] == 1]
negative_samples = data[data['flag'] == 0]

# 下采样负类别样本
num_samples = len(negative_samples)
positive_samples_downsampled = resample(positive_samples, replace=False, n_samples=num_samples, random_state=42)

# 合并下采样后的样本
downsampled_data = pd.concat([negative_samples, positive_samples_downsampled])
```

```python
downsampled_data.reindex()
```

```python
downsampled_data['flag'].value_counts()
```

```python
x = downsampled_data.copy()
y = downsampled_data['flag']
del x['flag']
```

```python
del x['source_city_name']
del x['target_city_name']
```

```python
model = xgb.XGBClassifier(learning_rate=0.1, n_estimators=100, max_depth=3)
model.fit(x, y)
```

```python
scores = cross_val_score(model, X=x, y=y, verbose=0, cv = 10, scoring='f1')
scores
```
